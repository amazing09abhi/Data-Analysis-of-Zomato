```python
data_country= pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Country-Code.xlsx')
data_country.head()
```

```python
final_data=pd.merge(data,data_country, on='Country Code', how= 'left' )
```

```python
country_names =final_data.Country.value_counts().index
```

```python
country_val= final_data.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct="'%1.2f%%'")
```

**Observation**:Zomato is earning mainly in India.
